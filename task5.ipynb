```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('accidents.csv')
```

```python
print("Sample data:")
print(data.head())
```

```python
print("\nDataset summary:")
print(data.info())
```

```python
print("\nAccident count statistics:")
print(data.describe())
```

```python
time_periods = ['0-3 hrs. (Night)', '3-6 hrs. (Night)', '6-9 hrs (Day)', '9-12 hrs (Day)',
                '12-15 hrs (Day)', '15-18 hrs (Day)', '18-21 hrs (Night)', '21-24 hrs (Night)']
```

```python
plt.figure(figsize=(12, 8))
```

```python
for i, time_period in enumerate(time_periods):
    plt.subplot(2, 4, i+1)
    sns.barplot(data=data, x='YEAR', y=time_period, palette='coolwarm')
    plt.title(f'Accidents - {time_period}')
    plt.xlabel('Year')
    plt.ylabel('Accident Count')
    plt.xticks(rotation=45)
```

```python
plt.tight_layout()
plt.suptitle('Accident Counts by Time Periods for Each Year', y=1.02)
plt.show()
```

```python
data['Total'] = data[time_periods].sum(axis=1)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='YEAR', y='Total', marker='o', color='b')
plt.title('Total Accidents Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Accident Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```
